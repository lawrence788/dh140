```python
import requests
import re
import pandas as pd
import nltk
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
from nltk.probability import FreqDist
from nltk.stem.lancaster import LancasterStemmer
from scipy import stats
import networkx as nx
import matplotlib.pyplot as plt

    
#get the text for Romeo and Juliet
response = requests.get("http://shakespeare.mit.edu/romeo_juliet/full.html")
document = BeautifulSoup(response.text, "html.parser")
text = ""
for doc in document.find_all(['blockquote', 'h3']):
    text += (doc.text.strip())
    text += (" ")

#print(text)

characters = [
    'ROMEO',
    'JULIET',
    'TYBALT',
    'MERCUTIO',
    'BENVOLIO',
    'CAPULET',
    'PARIS',
    'FRIAR LAURENCE',
    'MONTAGUE',
    'ESCALUS',
    'NURSE',
    'FRIAR JOHN',
    'MUSICIAN',
    'GUARD',
    'POTPAN',
    'ANTHONY'
]


acts = text.split('ACT')[:]
#print(len(acts))
#for act in acts:
    #print(act)
#    print(len(act.split('SCENE')))

charnum = {}
connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    charnum[k] = 0

for i in acts:
    for j in i.split('SCENE')[:]:
        scenechars = []
        for k in characters:
            if j.find(k) != -1:
                scenechars.append(k)
                charnum[k] += 1
        for a in range(len(scenechars)-1):
            for b in range(a+1,len(scenechars)):
                connections[(scenechars[a],scenechars[b])] += 1
            
print(charnum)
print(connections)

nxgraphitems = []

# for k,v in charnum.items():
#     dashnode = k
#     graphitems.append(dashnode)

for k,v in connections.items():
    if v != 0:
        dashedge = (k[0],k[1])
        nxgraphitems.append(dashedge)
    
G = nx.Graph()
G.add_edges_from(nxgraphitems)

dc = nx.degree_centrality(G)
for i in sorted(dc, key=dc.get, reverse=True):
    print('{:15s}: {:.3f}'.format(i.title(), dc[i]))
    
from networkx.algorithms.community.centrality import girvan_newman
communities = girvan_newman(G)

node_groups = []
for com in next(communities):
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

The information contained in the graph represents how many scenes are shared between two nodes. For example, Romeo and Juliet share four scenes, and Romeo and Mercutio share 4 scenes as well. These are represented as the weight on the edges. 

Also, the graph has tried to create groups using the nodes, and we can see that although most of the characters
in blue are highly connected during the play, Friar John is a much more minor character and is thus classified
as the green group.

```python
nx.density(G)
```

As seen above, the network's density is 0.64. This mean that the play uses about 64% of the connections that 
are possible given the character set. This means that the play is small, but the characters are highly connected
between each other. 

The most important nodes are Romeo, Capulet, and Montague, as measured by degree centrality. All three have
degree centrality of 0.889. The second most important nodes are Benvolio, Paris, and Friar Laurence with 
degree centralities of 0.667. This makes sense, because Romeo is the main character of the play, and Capulet 
and Montague are the two main fueding families in the play.
